```python
# Dependencies and Setup
import pandas as pd

# File Load
school_csv = "Resources/schools_complete.csv"
student_csv = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_csv)
student_data = pd.read_csv(student_csv)

# Combine the data into a single dataset.  
district_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
district_data.tail()
```

```python
new_district_data = district_data.rename(columns={"Student ID" : "student_id", "School ID" : "school_id"})
new_district_data
```

```python
# Find total schools and total students

total_schools = 15      #school ids run from 0 to 14 consecutively 
total_students = 39170        #student ids run from 0 to 39169 consecutively


```

```python
# Find total school and district budget

huang_budget = 1910635
figueroa_budget = 1884411
shelton_budget = 1056600
hernandez_budget = 3022020
griffin_budget = 917500
wilson_budget = 1319574
cabrera_budget = 1081356
bailey_budget = 3124928
holden_budget = 248087
pena_budget = 585858
wright_budget = 1049400
rodriguez_budget = 254363
johnson_budget = 3094650
ford_budget = 1763916
thomas_budget = 1043130

total_district_budget = (huang_budget + figueroa_budget + shelton_budget + hernandez_budget + griffin_budget + wilson_budget + cabrera_budget + bailey_budget + holden_budget + pena_budget + wright_budget + rodriguez_budget + johnson_budget + ford_budget + thomas_budget)

#total_district_budget = 22356428
```

```python
# Find district average math score

#district_average_math_score = 78.98537145774827

district_average_math_score = new_district_data.math_score.mean()
```

```python
# Find district average reading score

#district_average_reading_score = 81.87784018381414

district_average_reading_score = new_district_data.reading_score.mean()
```

```python
# Find % passing math score

#new_district_data[["student_id", "school_name", "school_id", "math_score"]]  #sub dataframe

passing_math_score = len(new_district_data[new_district_data.math_score >= 70])

pct_district_passing_math = (passing_math_score / total_students) * 100

#pct_district_passing_math = 74.9808526933878
```

```python
# Find % passing reading score

passing_reading_score = len(new_district_data[new_district_data.reading_score >= 70])

pct_district_passing_reading = (passing_reading_score / total_students) * 100

#pct_district_passing_reading = 85.80546336482001        
```

```python
# Find % passing overall (reading and math)

passing_overall_score = len(new_district_data[(new_district_data.math_score >= 70) & (new_district_data.reading_score >= 70)]) 

pct_district_passing_overall = (passing_overall_score / total_students) * 100

#pct_district_passing_overall = 65.17232575950983

```

```python
district_summary = pd.DataFrame(
    {"Total Schools": [total_schools],
     "Total Students": [total_students],
     "Total Budget": [total_district_budget],
     "Average Math Score": [district_average_math_score],
     "Average Reading Score": [district_average_reading_score],
     "% Passing Math": [pct_district_passing_math],
     "% Passing Reading": [pct_district_passing_reading],
     "% Overall Passing": [pct_district_passing_overall]
    
    })
district_summary
```
